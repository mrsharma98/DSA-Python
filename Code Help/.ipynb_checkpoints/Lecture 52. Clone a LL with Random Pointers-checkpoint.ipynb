```python
# let's build a same LL for data and next
# now change the original and cloned LLs next pointer

# 1    2   3
# |  / | /
# v/   v/
# 1    2

# here origin's next will point to its cloned copy
# cloned next will point to the original next
# random is still pointing to same


# TC --> O(n)
# SC --> O(1)
```

```python
# 1. Create a clone list
# 2. Clone nodes add in between OG LL
# 3. copy random pointer
# here's how, original.next = original.random.next
# 4. remove the structure OR revert the changes made in step 2.
# 5. return clone LL head.
```

```python
class Node:
    def __init__(Self, val):
        self.val = val
        self.next = None
        self.random = None


cloneHead = None
cloneTail = None

def insertAtTail(head, tail, data):
    newNode = Node(data)
    if head == None:
        head = newNode
    else:
        tail.next = newNode
    tail = newNode
    return

def copyLL(head):
    
    # create a clone list
    temp = head
    while (temp != None):
        insertAtTail(cloneHead, cloneTail, data)
        temp = temp.next
        
    # Clone nodes add in between OG LL
    originalNode = head
    cloneNode = cloneHead
    
    while originalNode!=None and cloneNode!=None:
        nxt = originalNode.next
        originalNode.next = cloneNode
        originalNode = nxt
        
        nxt = cloneNode.next
        cloneNode.next = originalNode
        cloneNode = nxt
        
    
    # copy random pointer
    temp = head
    while temp != None:
        temp.next.random = temp.random.next if temp.random!=None else None
        temp = temp.next.next # so that temp goes from 1 to 1 to 2
    
    # remove the structure OR revert the changes made in step 2.
    originalNode = head
    cloneNode = cloneHead
    
    while originalNode != None and cloneNode != None:
        originalNode.next = cloneNode.next
        originalNode = originalNode.next
        
        if originalNode != None:
            cloneNode.next = originalNode.next
        cloneNode = cloneNode.next
    
    
    # return clone LL head
    return cloneHead
        
```
