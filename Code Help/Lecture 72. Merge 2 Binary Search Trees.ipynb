```python
def inorder(root, inorderVal):
    if root == None:
        return 
    
    inorder(root.left, inorderVal)
    inorderVal.append(root.val)
    inorder(root.right, inorderVal)


def mergeArrays(a, b):
    i = 0
    j = 0
    k = 0
    ans = []
    
    while i<len(a) and j<len(b):
        if (a[i]<b[j]):
            ans[k] = a[i]
            k+=1
            i+=1
        else:
            ans[k] = b[j]
            k+=1
            j+=1
            
    while i<len(a):
        ans[k] = a[i]
        i += 1
        k += 1
    
    while j<len(b):
        ans[k] = b[j]
        j += 1
        k += 1
        
    return ans


def inorderToBST(s, e, inorder):
    if (s>e):
        return None
    
    mid = (s+e)//2
    root = TreeNode(inorder[mid])
    root.left = inorderToBST(s, mid-1, inorder)
    root.right = inorderToBST(mid+1, e, inorder)
    return root
    
    
def mergeBST(root1, root2):
    # store inorder
    bst1 = []
    bst2 = []
    inorder(root1, bst1)
    inorder(root2, bst2)
    
    mergedArray = mergeArrays(bst1, bst2)
    
    start = 0
    end = len(mergedArray)
    return inorderToBST(start, end, mergedArray)
```

```python
# SC:- O(h1+h2)
def convertIntoSortedDLL(root, head):
    # base case
    if root == None:
        return None
    
    convertIntoSortedDLL(root.right, head)
    
    root.right = head
    if head!=None:
        head.left = root
    
    head = root
    
    convertIntoSortedDLL(root.left, head)
```

```python
def mergeSortedDLL(head1, head2):
    head = None
    tail = None
    
    while (head1!=None and head2!=None):
        if (head1.data < head2.data):
            if (head == None):
                head = head1
                tail = head1
                head1 = head1.right
            else:
                tail.right = head1
                head1.left = tail
                tail = head1
                head1 = head1.right
        else:
            if (head == None):
                head = head2
                tail = head2
                head2 = head2.right
            else:
                tail.right = head2
                head2.left = tail
                tail = head2
                head2 = head2.right
                
        while head1 != None:
            tail.right = head1
            head1.left = tail
            tail = head1
            head1 = head1.right
        
        while head2 != None:
            tail.right = head2
            head2.left = tail
            tail = head2
            head2 = head2.right
        
        return head
```

```python
def countNodes(head):
    cnt = 0
    temp = head
    while (temp!=None):
        cnt += 1
        temp = temp.right
    return cnt
```

```python
def DDLToBST(head, n):
    # base case
    if n<=0 head == None:
        return None
    
    left = DDLtoBST(head, n//2)
    root = head
    root.left = left
    
    head = head.right
    root.right = DDLToBST(head, n-n//2-1)
    
    return root
```

```python
def mergeBST(root1, root2):
    head1 = None
    head2 = None
    convertIntoSortedDLL(root1, head1)
    convertIntoSortedDLL(root2, head2)
    
    head = mergeSortedDLL(head1, head2)
    noOfNodes = countNodes(head)
    
    return DDLToBST(head, noOfNodes)
```
