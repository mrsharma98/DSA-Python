```python
# Approach

# step-1:- node to parent mapping
# map<Node*, Node*> nodeToParent mapping
# using Level over traversal
# O(n)

# step-2:- find the target node -- O(h)

# step-3:- mapping for visited node
# map <Node*, bool>

# use -- queue, push the target node
# there are 3 possibility -- left, right, parent
```

```python
# TC:- O(n) + O(n) = O(n) or O(nlogn)
# SC:- O(n)

class Solution:
    def createParentMapping(self, root, target, nodeToParent):
        res = None
        q = []
        q.append(root)
        nodeToParent[root] = None
        
        while (len(q)>0):
            front = q[0]
            
            if front.data == target:
                res = front
            
            if front.left != None:
                nodeToParent[front.left] = front
                q.append(front.left)
            if front.right != None:
                nodeToParent[front.right] = front
                q.append(front.right)
        
        return res
    
    
    def burnTree(self, root, nodeToParent):
        visited = {}
        q = []
        q.append(root)
        visited[root] = true
        ans = 0
        
        while (len(q)>0):
            flag = 0
            size = len(q)
            for i in range(size):
                # process neighbouring nodes
                front = q[0]
                q.pop(0)
                
                if (front.left!=None and not visited[front.left]):
                    flag = 1
                    q.append(front.left)
                    visited[front.left] = 1
                if (front.right!=None and not visited[front.right]):
                    flag = 1
                    q.append(front.right)
                    visited[front.right] = 1
                    
                if (nodeToParent[front] and not visited[nodeToParent]):
                    flag = 1
                    q.append(nodeToParent[front])
                    visited[nodeToParent[front]] = 1
                    
            if flag == 1:
                ans += 1
        return ans
    
    
    def minTime(self, root, target):
        nodeToParent = {}
        targetNode = createParentMapping(root, target, nodeToParent)
        
        ans = burnTree(targetNode, nodeToParent)
        
        return ans
    
    
```
